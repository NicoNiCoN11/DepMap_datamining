## Find co-subunits of the complex

```python
saga_ls = ['ADA','ATXN7L1','ATXN7L2','ATXN7L3','KAT2A','SGF29','SUPT7L','TAF5L','TAF6L','SUPT20H',
           'TAF10','TAF12','TAF9L','SUPT3H','TADA2B','TADA3','USP22']
atac_ls = ['KAT14','KAT2A','MBIP','DR1','SGF29','TADA2A','TADA3','WDR5','YEATS2','ZZZ3','CSRP2BP']
```

```python
combine_ls = list(set(saga_ls)|set(atac_ls))
```

```python
# 导入personr计算两个变量之间的皮尔逊系数
from scipy.stats import pearsonr
```

```python
# 对df_crispr这个dataframe进行转置，找出其中combine_ls中包含的列，去掉任何有缺失值的行，再转置回来，得到一个新的数据框df_crispr_com
df_crispr_com = df_crispr.T.loc[[c for c in combine_ls if c in df_crispr.columns],:].dropna(how="any").T
```

```python
df_crispr_com
```

```python
# 导入scipy库中的pearsonr函数，用来计算两个变量之间的皮尔逊相关系数
from scipy.stats import pearsonr

# 对df_crispr这个数据框进行转置，筛选出combine_ls中包含的列，去掉任何有缺失值的行，再转置回来，得到一个新的数据框df_crispr_com
df_crispr_com = df_crispr.T.loc[[c for c in combine_ls if c in df_crispr.columns],:].dropna(how='any').T

# 创建一个空的字典，用来存储不同基因之间的相关系数
cor_dic = {}
# 开始一个循环，遍历df_crispr_com的所有列，每一列代表一个基因
for i in df_crispr_com.columns:
    # 创建一个空的列表，用来存储当前基因与其他基因的相关系数
    cor_ls = []
    # 开始一个内层循环，遍历df_crispr_com的所有列，每一列代表一个基因
    for j in df_crispr_com.columns:
        # 调用pearsonr函数，计算当前基因i和基因j之间的相关系数，取第一个元素（第二个元素是p值），并把它添加到cor_ls列表中
        cor_ls.append(pearsonr(df_crispr_com[i],df_crispr_com[j])[0])   
    # 把cor_ls列表赋值给cor_dic字典的键i，即当前基因的名称
    cor_dic[i] = cor_ls
# 把cor_dic字典转换成一个数据框，列名和行名都是df_crispr_com的列名，即基因的名称
data = pd.DataFrame(cor_dic,columns=df_crispr_com.columns)
# 把数据框的行名设置为df_crispr_com的列名，即基因的名称
data.index = df_crispr_com.columns
# 打印数据框，显示不同基因之间的相关系数矩阵
print(data)
```

```python
# 导入networkx库，用来创建和操作网络图
import networkx as nx
# 导入numpy库，用来进行数值计算
import numpy as np

# 设置随机数种子，用来保证每次运行代码得到相同的结果
np.random.seed(1027)

# 把数据框转换成一个长格式的数据框，每一行包含两个基因的名称和它们之间的相关系数，命名为links
links = data.stack().reset_index()
# 给links数据框的列名重新命名为gene1, gene2, corr，分别表示第一个基因，第二个基因，和相关系数
links.columns = ['gene1','gene2','corr']
print(links)
```

```python
# 对links数据框进行筛选，只保留绝对值大于0.2且不是自身的相关系数，命名为links_filtered
links_filtered=links.loc[(abs(links['corr']) > 0.1) & (links['gene1'] != links['gene2']) ]
print (links_filtered)
```

```python
# 创建一个图形对象，设置大小为8*4英寸，分辨率为300像素每英寸
plt.figure(figsize=(16,8),dpi=300)

# 把links_filtered数据框转换成一个网络图对象，每一行代表一条边，起点是gene1，终点是gene2，权重是corr
G=nx.from_pandas_edgelist(links_filtered, 'gene1', 'gene2',['corr'])
# 创建一个空的列表，用来存储每条边的颜色
color_list = []
# 开始一个循环，遍历网络图的所有边，每条边由两个节点u和v组成
for u,v in G.edges():
    # 把当前边的权重，即相关系数，添加到color_list列表中
    color_list.append(G[u][v]['corr'])
    
print (color_list)
```

```python
# 用spring_layout算法为网络图的每个节点分配一个位置，命名为pos
pos = nx.spring_layout(G)

# 绘制网络图，使用pos作为节点的位置，显示节点的标签，设置节点的颜色为橙色，大小为600，边的颜色为浅蓝色，宽度为3.0，标签的字体大小为6
nx.draw(G, pos,with_labels=True, node_color='orange', node_size=600, edge_color='lightblue',
        width=3.0,font_size=6)

# 显示图形
plt.show()
```
