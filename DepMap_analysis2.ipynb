## Find co-subunits of the complex

```python
saga_ls = ['ADA','ATXN7L1','ATXN7L2','ATXN7L3','KAT2A','SGF29','SUPT7L','TAF5L','TAF6L','SUPT20H',
           'TAF10','TAF12','TAF9L','SUPT3H','TADA2B','TADA3','USP22']
atac_ls = ['KAT14','KAT2A','MBIP','DR1','SGF29','TADA2A','TADA3','WDR5','YEATS2','ZZZ3','CSRP2BP']
```

```python
combine_ls = list(set(saga_ls)|set(atac_ls))
```

```python
# å¯¼å…¥personrè®¡ç®—ä¸¤ä¸ªå˜é‡ä¹‹é—´çš„çš®å°”é€Šç³»æ•°
from scipy.stats import pearsonr
```

```python
# å¯¹df_crisprè¿™ä¸ªdataframeè¿›è¡Œè½¬ç½®ï¼Œæ‰¾å‡ºå…¶ä¸­combine_lsä¸­åŒ…å«çš„åˆ—ï¼ŒåŽ»æŽ‰ä»»ä½•æœ‰ç¼ºå¤±å€¼çš„è¡Œï¼Œå†è½¬ç½®å›žæ¥ï¼Œå¾—åˆ°ä¸€ä¸ªæ–°çš„æ•°æ®æ¡†df_crispr_com
df_crispr_com = df_crispr.T.loc[[c for c in combine_ls if c in df_crispr.columns],:].dropna(how="any").T
```

```python
df_crispr_com
```

```python
# å¯¼å…¥scipyåº“ä¸­çš„pearsonrå‡½æ•°ï¼Œç”¨æ¥è®¡ç®—ä¸¤ä¸ªå˜é‡ä¹‹é—´çš„çš®å°”é€Šç›¸å…³ç³»æ•°
from scipy.stats import pearsonr

# å¯¹df_crisprè¿™ä¸ªæ•°æ®æ¡†è¿›è¡Œè½¬ç½®ï¼Œç­›é€‰å‡ºcombine_lsä¸­åŒ…å«çš„åˆ—ï¼ŒåŽ»æŽ‰ä»»ä½•æœ‰ç¼ºå¤±å€¼çš„è¡Œï¼Œå†è½¬ç½®å›žæ¥ï¼Œå¾—åˆ°ä¸€ä¸ªæ–°çš„æ•°æ®æ¡†df_crispr_com
df_crispr_com = df_crispr.T.loc[[c for c in combine_ls if c in df_crispr.columns],:].dropna(how='any').T

# åˆ›å»ºä¸€ä¸ªç©ºçš„å­—å…¸ï¼Œç”¨æ¥å­˜å‚¨ä¸åŒåŸºå› ä¹‹é—´çš„ç›¸å…³ç³»æ•°
cor_dic = {}
# å¼€å§‹ä¸€ä¸ªå¾ªçŽ¯ï¼ŒéåŽ†df_crispr_comçš„æ‰€æœ‰åˆ—ï¼Œæ¯ä¸€åˆ—ä»£è¡¨ä¸€ä¸ªåŸºå› 
for i in df_crispr_com.columns:
    # åˆ›å»ºä¸€ä¸ªç©ºçš„åˆ—è¡¨ï¼Œç”¨æ¥å­˜å‚¨å½“å‰åŸºå› ä¸Žå…¶ä»–åŸºå› çš„ç›¸å…³ç³»æ•°
    cor_ls = []
    # å¼€å§‹ä¸€ä¸ªå†…å±‚å¾ªçŽ¯ï¼ŒéåŽ†df_crispr_comçš„æ‰€æœ‰åˆ—ï¼Œæ¯ä¸€åˆ—ä»£è¡¨ä¸€ä¸ªåŸºå› 
    for j in df_crispr_com.columns:
        # è°ƒç”¨pearsonrå‡½æ•°ï¼Œè®¡ç®—å½“å‰åŸºå› iå’ŒåŸºå› jä¹‹é—´çš„ç›¸å…³ç³»æ•°ï¼Œå–ç¬¬ä¸€ä¸ªå…ƒç´ ï¼ˆç¬¬äºŒä¸ªå…ƒç´ æ˜¯på€¼ï¼‰ï¼Œå¹¶æŠŠå®ƒæ·»åŠ åˆ°cor_lsåˆ—è¡¨ä¸­
        cor_ls.append(pearsonr(df_crispr_com[i],df_crispr_com[j])[0])   
    # æŠŠcor_lsåˆ—è¡¨èµ‹å€¼ç»™cor_dicå­—å…¸çš„é”®iï¼Œå³å½“å‰åŸºå› çš„åç§°
    cor_dic[i] = cor_ls
# æŠŠcor_dicå­—å…¸è½¬æ¢æˆä¸€ä¸ªæ•°æ®æ¡†ï¼Œåˆ—åå’Œè¡Œåéƒ½æ˜¯df_crispr_comçš„åˆ—åï¼Œå³åŸºå› çš„åç§°
data = pd.DataFrame(cor_dic,columns=df_crispr_com.columns)
# æŠŠæ•°æ®æ¡†çš„è¡Œåè®¾ç½®ä¸ºdf_crispr_comçš„åˆ—åï¼Œå³åŸºå› çš„åç§°
data.index = df_crispr_com.columns
# æ‰“å°æ•°æ®æ¡†ï¼Œæ˜¾ç¤ºä¸åŒåŸºå› ä¹‹é—´çš„ç›¸å…³ç³»æ•°çŸ©é˜µ
print(data)
```

```python
# å¯¼å…¥networkxåº“ï¼Œç”¨æ¥åˆ›å»ºå’Œæ“ä½œç½‘ç»œå›¾
import networkx as nx
# å¯¼å…¥numpyåº“ï¼Œç”¨æ¥è¿›è¡Œæ•°å€¼è®¡ç®—
import numpy as np

# è®¾ç½®éšæœºæ•°ç§å­ï¼Œç”¨æ¥ä¿è¯æ¯æ¬¡è¿è¡Œä»£ç å¾—åˆ°ç›¸åŒçš„ç»“æžœ
np.random.seed(1027)

# æŠŠæ•°æ®æ¡†è½¬æ¢æˆä¸€ä¸ªé•¿æ ¼å¼çš„æ•°æ®æ¡†ï¼Œæ¯ä¸€è¡ŒåŒ…å«ä¸¤ä¸ªåŸºå› çš„åç§°å’Œå®ƒä»¬ä¹‹é—´çš„ç›¸å…³ç³»æ•°ï¼Œå‘½åä¸ºlinks
links = data.stack().reset_index()
# ç»™linksæ•°æ®æ¡†çš„åˆ—åé‡æ–°å‘½åä¸ºgene1, gene2, corrï¼Œåˆ†åˆ«è¡¨ç¤ºç¬¬ä¸€ä¸ªåŸºå› ï¼Œç¬¬äºŒä¸ªåŸºå› ï¼Œå’Œç›¸å…³ç³»æ•°
links.columns = ['gene1','gene2','corr']
print(links)
```

```python
# å¯¹linksæ•°æ®æ¡†è¿›è¡Œç­›é€‰ï¼Œåªä¿ç•™ç»å¯¹å€¼å¤§äºŽ0.2ä¸”ä¸æ˜¯è‡ªèº«çš„ç›¸å…³ç³»æ•°ï¼Œå‘½åä¸ºlinks_filtered
links_filtered=links.loc[(abs(links['corr']) > 0.1) & (links['gene1'] != links['gene2']) ]
print (links_filtered)
```

```python
# åˆ›å»ºä¸€ä¸ªå›¾å½¢å¯¹è±¡ï¼Œè®¾ç½®å¤§å°ä¸º8*4è‹±å¯¸ï¼Œåˆ†è¾¨çŽ‡ä¸º300åƒç´ æ¯è‹±å¯¸
plt.figure(figsize=(16,8),dpi=300)

# æŠŠlinks_filteredæ•°æ®æ¡†è½¬æ¢æˆä¸€ä¸ªç½‘ç»œå›¾å¯¹è±¡ï¼Œæ¯ä¸€è¡Œä»£è¡¨ä¸€æ¡è¾¹ï¼Œèµ·ç‚¹æ˜¯gene1ï¼Œç»ˆç‚¹æ˜¯gene2ï¼Œæƒé‡æ˜¯corr
G=nx.from_pandas_edgelist(links_filtered, 'gene1', 'gene2',['corr'])
# åˆ›å»ºä¸€ä¸ªç©ºçš„åˆ—è¡¨ï¼Œç”¨æ¥å­˜å‚¨æ¯æ¡è¾¹çš„é¢œè‰²
color_list = []
# å¼€å§‹ä¸€ä¸ªå¾ªçŽ¯ï¼ŒéåŽ†ç½‘ç»œå›¾çš„æ‰€æœ‰è¾¹ï¼Œæ¯æ¡è¾¹ç”±ä¸¤ä¸ªèŠ‚ç‚¹uå’Œvç»„æˆ
for u,v in G.edges():
    # æŠŠå½“å‰è¾¹çš„æƒé‡ï¼Œå³ç›¸å…³ç³»æ•°ï¼Œæ·»åŠ åˆ°color_liståˆ—è¡¨ä¸­
    color_list.append(G[u][v]['corr'])
    
print (color_list)
```

```python
# ç”¨spring_layoutç®—æ³•ä¸ºç½‘ç»œå›¾çš„æ¯ä¸ªèŠ‚ç‚¹åˆ†é…ä¸€ä¸ªä½ç½®ï¼Œå‘½åä¸ºpos
pos = nx.spring_layout(G)

# ç»˜åˆ¶ç½‘ç»œå›¾ï¼Œä½¿ç”¨posä½œä¸ºèŠ‚ç‚¹çš„ä½ç½®ï¼Œæ˜¾ç¤ºèŠ‚ç‚¹çš„æ ‡ç­¾ï¼Œè®¾ç½®èŠ‚ç‚¹çš„é¢œè‰²ä¸ºæ©™è‰²ï¼Œå¤§å°ä¸º600ï¼Œè¾¹çš„é¢œè‰²ä¸ºæµ…è“è‰²ï¼Œå®½åº¦ä¸º3.0ï¼Œæ ‡ç­¾çš„å­—ä½“å¤§å°ä¸º6
nx.draw(G, pos,with_labels=True, node_color='orange', node_size=600, edge_color='lightblue',
        width=3.0,font_size=6)

# æ˜¾ç¤ºå›¾å½¢
plt.show()
```
