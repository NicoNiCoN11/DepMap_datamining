## Add model to classify the cell types

```python
df_cell = pd.read_csv("Model.csv")
df_cell
```

```python
set(df_cell['OncotreeLineage'])
```

```python
# 创建一个没有重复肿瘤类型的集合
cancer_ls = list(set(df_cell["OncotreeLineage"].dropna()))
```

```python
# 计算有多少个元素并为每个元素对应一个颜色
import matplotlib.colors as mclors
import matplotlib.pyplot as plt

# 计算列表中有多少元素
n = len(cancer_ls)
cmap = plt.get_cmap("rainbow")  # 还有其他颜色图"rainbow", "viridis", "plasma"等
color_list = [cmap(i/n) for i in range(n)]
```

```python
gene = "SOX10"
plt.figure(figsize=(6,4),dpi=300)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.rcParams["font.size"] = 8
plt.rcParams["font.family"] = "Helvetica"

color_ind = 0
for cancertype in cancer_ls:
    df_cancer = df_cell[df_cell['OncotreeLineage']==cancertype]
    cell_ls = [c for c in list(df_cancer['ModelID']) if c in ind_ls]
    plt.scatter(df_expression.loc[cell_ls, gene], df_crispr.loc[cell_ls, gene], s=6, alpha=0.5, label=cancertype)
    color_ind += 1

plt.xlabel('Expression of %s'%gene)
plt.ylabel('CRISPR score of %s'%gene)
plt.legend(fontsize='xx-small',loc='lower left',ncol=2)

plt.show()
```

## Assocaition between gene mutation and essentiality

```python
mut_df = pd.read_csv('OmicsSomaticMutations.csv')
```

```python
mut_df.index = list(mut_df['ModelID'])
```

```python
mut_df['MutationType'] = mut_df['HugoSymbol'] + '_' + mut_df['ProteinChange']
```

```python
mut_df
```

```python
## Get cell lines that are common in CRISPR and mutation data

ind_mut = list(set(df_crispr.index)&set(mut_df.index))
```

```python
gene = 'KRAS'
mut = "G12D"  # 定义基因突变的变量
muttype = gene + '_p.' + mut  # 拼接基因和突变的变量，得到突变类型的变量
# 从数据框mut_df中获取突变类型为muttype这一列，然后选取“ModelID”这一列，得到发生了muttpye类型突变的细胞，并整理到mut_cell列表中
mut_cell = mut_df[mut_df["MutationType"]==muttype]['ModelID']  
# 为数据框添加新的一列“MutLabel”，如果细胞在mut_cell中则赋值1，否则为0
df_crispr['MutLabel'] = [1 if c in mut_cell else 0 for c in df_crispr.index]
# 创建一个新的图形，设置图形的大小和分辨率
plt.figure(figsize=(3,2),dpi=300)
# 获取当前的坐标轴对象
ax = plt.gca()
# 隐藏顶部和右侧的轴线
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 设置字体的大小和类型
plt.rcParams["font.size"] = 8
plt.rcParams["font.family"] = "Helvetica"

# 从数据框df_crispr中选择那些发生的突变的那些行，然后选择之前定义的gene这一列，得到有突变细胞系的gene基因CRISPER Score
cs_ls1 = df_crispr[df_crispr['MutLabel']==1][gene]
# 从数据框中选择那些没有发生突变的行，选择gene这一列，没发生突变的细胞系的gene基因CRISPR score
cs_ls2 = df_crispr[df_crispr['MutLabel']==0][gene]

# 绘制一个箱线图，显示两个序列的分布情况，设置标签和是否显示异常值
plt.boxplot([cs_ls1,cs_ls2],labels=['WithMut','WitoutMut'],showfliers=False)
# 设置y轴的标签
plt.ylabel('CRISPR score for %s'%gene)
# 保存图形为一个png文件，设置分辨率、背景色和边距
plt.savefig(f"{gene}"'_Boxplot.png',dpi=300,facecolor=None,bbox_inches='tight')
# 显示图形
plt.show()
```

```python
plt.figure(figsize=(60,32),dpi=300)
plt.rcParams["font.size"] = 8
plt.rcParams["font.family"] = "Helvetica"
# 分析具体在哪些cancertype中此类突变重要
# df_cell
n = 1
for cancertype in cancer_ls:
    # 从数据框df_cell中选取肿瘤类型为cancertype的那些行，得到一个新的数据框df_cancer
    df_cancer = df_cell[df_cell["OncotreeLineage"]==cancertype]
    # 从数据框df_cancer中选取“ModelID”这一列，得到一个序列，然后将其转换为一个列表，再筛选出那些在数据框df_crispr的行标签（index）中的细胞系（cell line），得到一个列表cell_ls
    cell_ls = [c for c in list(df_cancer['ModelID']) if c in df_crispr.index]
    df_sel = df_crispr.loc[cell_ls,:]
    
    # 从数据框df_crispr中选择那些发生的突变的那些行，然后选择之前定义的gene这一列，得到有突变细胞系的gene基因CRISPER Score
    cs_ls1 = df_sel[df_sel['MutLabel']==1][gene]
    # 从数据框中选择那些没有发生突变的行，选择gene这一列，没发生突变的细胞系的gene基因CRISPR score
    cs_ls2 = df_sel[df_sel['MutLabel']==0][gene]
    
    plt.subplot(7,5,n)  #五行七列的子图 
    plt.boxplot([cs_ls1,cs_ls2],labels=['WithMut','WitoutMut'],showfliers=False)
    plt.ylabel('CRISPR score in %s'%cancertype)
    n += 1
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```

```python
gene = 'BRAF'
cancertype = 'Skin'
plt.figure(figsize=(3,2),dpi=300)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.rcParams["font.size"] = 8
plt.rcParams["font.family"] = "Helvetica"

plt.scatter(copynum_df.loc[ind_cn, gene], df_crispr.loc[ind_cn, gene],s=10, alpha=0.5, color='blue',label='All')
df_cancer = df_cell[df_cell["OncotreeLineage"] == cancertype]
cell_ls = [c for c in list(df_cancer['ModelID']) if c in ind_cn]
plt.scatter(copynum_df.loc[cell_ls, gene], df_crispr.loc[cell_ls, gene], s=6, alpha=0.5, color='red', label=cancertype)
plt.xlabel('Copy number of %s'%gene)
plt.ylabel('CRISPR score of %s'%gene)
plt.legend(fontsize='x-small',loc='lower right')

plt.show()
```
